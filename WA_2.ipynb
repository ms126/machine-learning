# Q2. Computational Graph + Back-propagation

$x = [1, 0], y = 1, w = [-1, 2]$

Forward pass: $z=w \cdot x=(-1)(1) + (2)(0)=-1$, $\hat{y}=sigmoid(-1)=\frac{1}{1+e^{-(-1)}} \approx 0.26894142137$, prediction loss $=(sigmoid(-1) - 1)^2=(0.26894142137-1)^2 \approx 0.534446646$

Backward pass:

$\frac{\partial loss}{\partial\hat{y}}=2(\hat{y}-y)=2(0.26894142137 - 1)=-1.46211715$, $\frac{\partial sigmoid(z)}{\partial z}=sigmoid(-1) \cdot 1-(sigmoid(-1))\approx -0.19661193324$

$\frac{\partial loss}{\partial z}=\frac{\partial loss}{\partial\hat{y}} \cdot \frac{\partial sigmoid(z)}{\partial z}=1.46211715 \cdot -0.19661193324 \approx -0.287469679$

$\frac{\partial loss}{\partial w_1}=\frac{\partial loss}{\partial z} \cdot x_1 = 0.287469679 \cdot 1 = -0.287469679$ and $\frac{\partial loss}{\partial w_2}=\frac{\partial loss}{\partial z} \cdot x_2 = -0.287469679 \cdot 0 = 0 \rightarrow [-0.287, 0]$

# Q3. Convolutional Neural Network

stride = $2$, zero-padding = $1$, output size = $\frac{W-K+2P}{S}+1 = \frac{7-3+2(1)}{2}+1 = 4 \rightarrow 4x4$ matrix

```python
import numpy as np

image = np.array([[3, 4, 2, 5, 6, 2, 3],
                  [1, 7, 2, 3, 3, 6, 7],
                  [2, 5, 3, 6, 5, 5, 6],
                  [2, 4, 4, 7, 6, 7, 5],
                  [3, 6, 3, 6, 7, 8, 6],
                  [4, 5, 5, 5, 7, 6, 7],
                  [6, 4, 3, 6, 6, 7, 7]])

filter_matrix = np.array([[0, -1, 0],
                          [-1, 5, -1],
                          [0, -1, 0]])

stride = 2
padding = 1

# Zero-padding
image_padded = np.pad(image, padding, mode='constant')

# Output size
output_size = (image.shape[0] - filter_matrix.shape[0] + 2 * padding) // stride + 1

# Output matrix
output_convolution = np.zeros((output_size, output_size))

# Convolution
for i in range(0, image_padded.shape[0] - filter_matrix.shape[0] + 1, stride):
    for j in range(0, image_padded.shape[1] - filter_matrix.shape[1] + 1, stride):
        output_convolution[i // stride, j // stride] = np.sum(image_padded[i:i+filter_matrix.shape[0], j:j+filter_matrix.shape[1]] * filter_matrix)

print("Resulting output after convolution = ")
print(output_convolution)

# Max pooling
pool_filter_size = 4
pool_stride = 1

# Calculate output size after max pooling
output_pooling_size = (image.shape[0] - pool_filter_size) // pool_stride + 1

# Initialize output matrix for max pooling
output_pooling = np.zeros((output_pooling_size, output_pooling_size))

# Max pooling operation
for i in range(0, image.shape[0] - pool_filter_size + 1, pool_stride):
    for j in range(0, image.shape[1] - pool_filter_size + 1, pool_stride):
        output_pooling[i // pool_stride, j // pool_stride] = np.max(image[i:i+pool_filter_size, j:j+pool_filter_size])

print("\nResulting output after max pooling = ")
print(output_pooling)
```
