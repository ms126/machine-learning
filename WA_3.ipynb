# Q2. Ensemble Learning

### **Similarities:**

The bagging and random forest methods both belong to the ensemble learning family. The ensemble learning family is where multiple models are trained to solve the same issue, and they are then combined to enhance performance. Both of the methods introduce randomness to reduce variance in the model and prevent overfitting of the dataset.

### **Differences:**

Bagging incorporates randomness through sampling with replacememt, whereas random forest uses a subset of features for each tree, addresses the issue of correlated trees. Bagging is not restricted by the amount of trees, but random forest typically uses many more (hundreds-thousands) to achieve a stronger accuracy.

### Weight vector and maximum margin: 

$x \cdot w + b = 0$, $w = (a, b)^T \rightarrow ax_1 + bx_2 - \frac{3}{2} = 0$ ... $a = 1, b = 1$ **and** $w = (1, 1)^T$

maximum margin hyper-plane should have a slope of -1 *and* $x_1 = \frac{3}{2}, x_2 = 0$ **so** the support vector is $(\frac{3}{2}, 0)$

maximum margin = $y_i(\beta_0+\beta_1x_{i1}+...+\beta_px_{ip})≥M$ $\forall i = 1,..,n$
... $M=\frac{1}{||w||} = \frac{1}{\sqrt2} \approx 0.707106781$

If you remove one of the support vectors in this dataset, the size of the maximum margin will increase. This is because removing these vectors will shift the position of our hyperplane, causing it to adjust to a new dataset, which has a larger separation.
